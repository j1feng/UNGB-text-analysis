```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
speech = pd.read_csv('UNGDspeeches.csv')
```

```python
speech.head(5)
```

## Distribution over the years

Number of Speech vs. Year

```python
speech.groupby('year')['doc_id'].agg({'count'}).head(5)
```

```python
speech.groupby('year')['doc_id'].agg({'count'}).plot()
plt.ylabel('Number of documents/speeches')
plt.title('Number of documents/speeches over the Years')
plt.show()
```

```python
speech.groupby('year')['country'].unique().apply(lambda x: len(x)).plot()
plt.ylabel('Number of countries')
plt.title('Number of countries in the United Nations over the Years')
plt.show()
```

Consistency check: one country has one document for each year
